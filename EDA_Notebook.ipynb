```python

outlier_col = ['Year_Birth' , 'Income', 'Recency', 'MntWines', 'MntFruits',
       'MntMeatProducts', 'MntFishProducts', 'MntSweetProducts',
       'MntGoldProds', 'NumDealsPurchases', 'NumWebPurchases',
       'NumCatalogPurchases', 'NumStorePurchases', 'NumWebVisitsMonth']
fig , axs = plt.subplots(5,3,figsize = (10,15))
axs = axs.flatten()
for col,ax in zip(outlier_col,axs):
    sns.boxplot(data = df , x = df[col] , ax = ax)
plt.tight_layout()
plt.show()
```

```python
# Remove the rows with outliers present in the 'Year_Birth' and 'Income' columns, as outliers in these columns can 
# significantly skew the results. The presence of outliers may adversely affect the imputation of null values, leading to 
# inaccurate predictions and misleading insights. By eliminating these outliers, we ensure that the data remains robust and
# reliable for subsequent analyses and imputation processes.
 
outlier_feat = ['Year_Birth' , 'Income']
def remove_outliers_iqr(df, columns):
  df_copy = df.copy()

  for col in columns:
    # Calculate the IQR
    q1 = df_copy[col].quantile(0.25)
    q3 = df_copy[col].quantile(0.75)
    iqr = q3 - q1

    # Determine outlier boundaries
    lower_bound = q1 - 1.5 * iqr
    upper_bound = q3 + 1.5 * iqr

    # Remove outliers
    df_copy = df_copy[(df_copy[col] >= lower_bound) & (df_copy[col] <= upper_bound)]

  return df_copy

df_cleaned = remove_outliers_iqr(df, outlier_feat)
df_cleaned
```

```python
# As Outliers are removed and we have to impute values in Income Column. Values are taken as the mean values of 
# the specific Education Category.
Mean_Master = df_cleaned[df_cleaned['Education'] == 'Master']['Income'].mean()
Mean_Graduation = df_cleaned[df_cleaned['Education'] == 'Graduation']['Income'].mean()
Mean_PhD = df_cleaned[df_cleaned['Education'] == 'PhD']['Income'].mean()
df_cleaned.loc[(df_cleaned['Income'] == 0) & (df_cleaned['Education'] == 'Masters'), 'Income'] = Mean_Master
df_cleaned.loc[(df_cleaned['Income'] == 0) & (df_cleaned['Education'] == 'Graduation'), 'Income'] = Mean_Graduation
df_cleaned.loc[(df_cleaned['Income'] == 0) & (df_cleaned['Education'] == 'PhD'), 'Income'] = Mean_PhD
```

```python
# Insert Age column with Year_Birth. Drop the Year_Birth
df_cleaned.insert(loc = 1 , column = 'Age' , value = 2024 - df_cleaned['Year_Birth'])
df_cleaned.drop(labels = 'Year_Birth' , axis = 1 , inplace = True)
```

```python
# New Variable "Revenue_Generated" in exchange of the sum of 6 product categories
df_cleaned.insert(loc = 9 , column = 'Revenue_generated' , value = df_cleaned['MntWines']+df_cleaned['MntFruits']
                  +df_cleaned['MntMeatProducts']+df_cleaned['MntFishProducts']+df_cleaned['MntSweetProducts']+df_cleaned['MntGoldProds'])
df_cleaned.drop(labels = [ 'MntWines', 'MntFruits','MntMeatProducts', 'MntFishProducts', 'MntSweetProducts',
       'MntGoldProds'] , axis = 1 , inplace = True)
```

```python
# Family variable as the sum of Kidhome + Teenhome + Marital_Status.Marital_Status can be imputed as 
# {'Divorced':1, 'Single':1, 'Married':2, 'Together':2, 'Widow':1, 'YOLO':1, 'Alone':1, 'Absurd':1}

dic_mar = {'Divorced':1 , 'Single':1 , 'Married':2 , 'Widow':1 , 'YOLO':1 , 'Absurd':1}
df_cleaned['Marital_Status'] = df_cleaned['Marital_Status'].replace(dic_mar)
df_cleaned.insert(loc = 3 , column = 'Family' , value = df_cleaned['Kidhome']+df_cleaned['Teenhome']
                  +df_cleaned['Marital_Status'])
df_cleaned.drop(labels = [ 'Kidhome', 'Teenhome','Marital_Status'] , axis = 1 , inplace = True)
```

```python
dic_edu = {'Graduation':16 , 'Master':18 , 'PhD':21 , 'Basic':12}
df_cleaned['Education'] = df_cleaned['Education'].replace(dic_edu)
df_cleaned = df_cleaned.rename(columns = {"Education":"Educational_Years"})
```

```python
# New Variable "Total_Purchases" by adding all purchase features
df_cleaned.insert(loc = 8 , column = 'Total_Purchases' , value = df_cleaned['NumDealsPurchases']+df_cleaned['NumWebPurchases']
                 +df_cleaned['NumCatalogPurchases']+df_cleaned['NumStorePurchases'])
df_cleaned.drop(labels = [ 'NumDealsPurchases', 'NumWebPurchases','NumCatalogPurchases', 'NumStorePurchases'] ,
                axis = 1 , inplace = True)
```

```python
# New Variable "TotalCampaignsAcc" as the total acceptance of advertising campaigns
df_cleaned.insert(loc = 10 , column = 'TotalCampaignsAcc' , value = df_cleaned['AcceptedCmp3']+df_cleaned['AcceptedCmp4']
                 +df_cleaned['AcceptedCmp5']+df_cleaned['AcceptedCmp1'] + df_cleaned['AcceptedCmp2'])
df_cleaned.drop(labels = ['AcceptedCmp3', 'AcceptedCmp4', 'AcceptedCmp5', 'AcceptedCmp1','AcceptedCmp2'] ,
                axis = 1 , inplace = True)
```
